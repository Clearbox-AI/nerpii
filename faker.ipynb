```python


faker = Faker()
loc_cols = [[i, list_entities[i]['entity']] for i in list_entities if list_entities[i] is not None]
        
person = [i[0] for i in loc_cols if i[1] == 'PERSON' and (('first' in i[0].lower()) and ('name' in i[0].lower()))]
```

```python
for col in person:
    synth[col] = synth.apply(lambda row: faker.first_name_female() if (row['first_name_gender'] == 'female' or row['first_name_gender'] == 'mostly_female') else row[col], axis=1)
    synth[col] = synth.apply(lambda row: faker.first_name_male() if (row['first_name_gender'] == 'male' or row['first_name_gender'] == 'mostly_male') else row[col], axis=1)
    synth[col] = synth.apply(lambda row: faker.first_name() if (row['first_name_gender'] == 'unknown' or row['first_name_gender'] == 'andy') else row[col], axis=1)
```

```python
from faker import Faker

if len(list_entities) > 0:
    try:
        faker = Faker()
        loc_cols = [[i, list_entities[i]['entity']] for i in list_entities if list_entities[i] is not None]
        
        adrs = [i[0] for i in loc_cols if i[1] == 'ADDRESS']
        for i in adrs:
            synth[i] = [faker.street_address() for i in range(synth.shape[0])]

        phone = [i[0] for i in loc_cols if i[1] == 'PHONE_NUMBER']
        for i in phone:
            synth[i] = [faker.phone_number() for i in range(synth.shape[0])]

        
    except:
        print('Something went wrong while create fake PIIs')
```
